```
import segmentation_models_pytorch as smp
import torch.nn as nn


class UNetChangeDetector(nn.Module):

    def __init__(self):

        super().__init__()

        self.model = smp.Unet(
            encoder_name="resnet34",
            encoder_weights="imagenet",
            in_channels=2,
            classes=1
        )

    def forward(self, x):

        return self.model(x)
```

```
import torch

from torch.utils.data import DataLoader

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print("Using Device:", DEVICE)

train_loader = DataLoader(
    train_dataset,
    batch_size=2,
    shuffle=True,
    num_workers=2
)

model = UNetChangeDetector().to(DEVICE)

criterion = torch.nn.BCEWithLogitsLoss()

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=1e-4
)

EPOCHS = 5
```

```
for epoch in range(EPOCHS):

    model.train()

    total_loss = 0

    for images, masks in train_loader:

        images = images.to(DEVICE)
        masks = masks.to(DEVICE)

        masks = masks.unsqueeze(1)

        outputs = model(images)

        loss = criterion(outputs, masks)

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS}")
    print(f"Loss: {total_loss / len(train_loader):.4f}")

torch.save(model.state_dict(), "best_model.pth")

print("Training Complete")
```

```
import os
import glob
import cv2
import rasterio
import numpy as np
import torch

from torch.utils.data import Dataset


class ChangeDetectionDataset(Dataset):

    def __init__(self, root_dir):

        self.pre_images = sorted(
            glob.glob(os.path.join(root_dir, "pre-event", "*.tif"))
        )

        self.post_images = sorted(
            glob.glob(os.path.join(root_dir, "post-event", "*.tif"))
        )

        self.targets = sorted(
            glob.glob(os.path.join(root_dir, "target", "*.tif"))
        )

    def __len__(self):
        return len(self.pre_images)

    def read_tif(self, path):

        with rasterio.open(path) as src:
            image = src.read(1)

        image = image.astype(np.float32)

        return image

    def __getitem__(self, idx):

        pre = self.read_tif(self.pre_images[idx])
        post = self.read_tif(self.post_images[idx])

        mask = self.read_tif(self.targets[idx])

        # Resize images
        pre = cv2.resize(pre, (512, 512))
        post = cv2.resize(post, (512, 512))
        mask = cv2.resize(     mask,     (512, 512),     interpolation=cv2.INTER_NEAREST )

        # Normalize images
        pre = pre / 255.0
        post = post / 255.0

        # Combine pre and post images
        image = np.stack([pre, post], axis=0)

        # CORRECT label remapping
        # 0 -> 0
        # 1 -> 0
        # 2 -> 1
        # 3 -> 1
        mask = np.where(mask >= 2, 1, 0).astype(np.float32)

        # Convert to tensors
        image = torch.tensor(image, dtype=torch.float32)
        mask = torch.tensor(mask, dtype=torch.float32)

        return image, mask
```
